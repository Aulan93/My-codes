```python
import random as rd
import numpy as np
import pandas as pd
from datetime import datetime
```

## We want to define the limits for hotness and coldness of each of the files.

Conditions: 
1. Number of access requests
2. Number of usages
3. Quality of the file
4. File type

Hot:
#1. > 0, $\quad$ 2.*1. $\geq$ 1, $\quad$  3. > 0 $\quad$ and 4. is B or S
else:
Cold


Create x number of files for training the algorithm, we can create three different datasets namely the training, testing and validation sets.  

Syntax

Define the number of possible options for the file type: pick one at random (B40, S40, L10, C10)
Number of access requests: range (0,x)
Number of usages: range (0,y)
Quality of the files: either 0 or 1 probaility of 80%

```python
acc_req=[] # access request
usg_req=[] # usage request
```

```python
def create_train_data(n):
  """It will help to create the training dataset"""
  rd.seed(101)
  ftype=[i for i in rd.choices(['B', 'S', 'L', 'C'], weights=[0.7,0.2,0.05,0.05], k=n)]
  quality=[i for i in rd.choices([1, 0] , weights=[0.8,0.2], k=n)]
  status=[0 for i in range(n)]
  for i in range(n):
      if ftype[i]=='B' or ftype[i]=='S':
          acc_req.append(np.random.randint(50))
          usg_req.append(np.random.randint(100))
      else:
          usg_req.append(0)
          acc_req.append(0)
      if acc_req[i]==0:
          usg_req[i]=0
  return pd.DataFrame(list(zip(ftype, quality, acc_req, usg_req, status)), columns =['File Type', 'Quality', 'Access Requests', 'Usage Requests', 'Status'])

def create_validation_data(n):
  """It will help to create the validation dataset"""
  rd.seed(101)
  ftype=[i for i in rd.choices(['B', 'S', 'L', 'C'], weights=[0.25,0.25,0.25,0.25], k=n)]
  quality=[i for i in rd.choices([1, 0] , weights=[0.5,0.5], k=n)]
  status=[0 for i in range(n)]
  for i in range(n):
      if ftype[i]=='B' or ftype[i]=='S':
          acc_req.append(np.random.randint(50))
          usg_req.append(np.random.randint(100))
      else:
          usg_req.append(0)
          acc_req.append(0)
      if acc_req[i]==0:
          usg_req[i]=0
  return pd.DataFrame(list(zip(ftype, quality, acc_req, usg_req, status)), columns =['File Type', 'Quality', 'Access Requests', 'Usage Requests', 'Status'])

def create_testing_data(n):
  """It will help to create the testing dataset"""
  rd.seed(101)
  ftype=[i for i in rd.choices(['B', 'S', 'L', 'C'], weights=[0.25,0.25,0.25,0.25], k=n)]
  quality=[i for i in rd.choices([1, 0] , weights=[0.5,0.5], k=n)]
  status=[0 for i in range(n)]
  for i in range(n):
      if ftype[i]=='B' or ftype[i]=='S':
          acc_req.append(np.random.randint(50))
          usg_req.append(np.random.randint(100))
      else:
          usg_req.append(0)
          acc_req.append(0)
      if acc_req[i]==0:
          usg_req[i]=0
  return pd.DataFrame(list(zip(ftype, quality, acc_req, usg_req, status)), columns =['File Type', 'Quality', 'Access Requests', 'Usage Requests', 'Status'])

def file_status(x):
  """It will return the status of the file: 1 for Hot file and  0 for Cold file."""
  if ((x['File Type']=='B') | (x['File Type']=='S')) & (x['Access Requests']>=1) & (x['Usage Requests']>=1) & (x['Quality']!=0):
      x['Status']=1
  else:
      x['Status']=0
  return x
```

```python
train= create_train_data(2450)
train= train.apply(file_status, axis=1)
train.to_csv('train.csv', index=False)

val=create_train_data(1050)
val=val.apply(file_status, axis=1)
val.to_csv('val.csv', index=False)


test=create_train_data(1500)
test=test.apply(file_status, axis=1)
test.to_csv('test.csv', index=False)
```
