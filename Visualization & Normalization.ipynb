```
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# Normalization of the Signal

```
#visualization

Gauss = pd.read_csv('gauss_data.csv')
plt.figure(figsize = (10,5))
plt.hist(Gauss['gauss'], color = 'green', bins = 100, label = 'Gaussian Signal')
#plt.xlim(100, 160)
plt.xlabel('Variation of the diphoton mass ($m_{\gamma\gamma}$)')
plt.ylabel('Number of Events')
plt.legend()
plt.show()
```

```
# normalization
(count, bin)=np.histogram(Gauss['gauss'], bins=100)

factor=0.24

plt.figure(figsize = (10,5))
plt.hist(bin[:-1], bin, weights=factor*count, label = 'Normalised Signal')
plt.xlim(100, 160)
plt.xlabel('Variation of the diphoton mass ($m_{\gamma\gamma}$)')
plt.ylabel('Number of Events')
plt.legend()
plt.show()
```

# Normalization of the exponential background

```
# visualization
Expo = pd.read_csv('expo_data.csv')

plt.figure(figsize = (10,5))
plt.hist(Expo['expo'], color = 'green', bins = 150, label = 'Exponential Background')
plt.xlabel('Variation of the diphoton mass ($m_{\gamma\gamma}$)')
plt.ylabel('Number of Events')
plt.legend()
plt.show()
```

```
# normalization
(counts, bins) = np.histogram(Expo['expo'], bins= 150)

factor = 2.57

plt.figure(figsize = (10,5))
plt.hist(bins[:-1], bins, weights=factor*counts, color = 'darkblue', label = 'Normalised Exponential Background')
plt.xlim(100, 160)
plt.xlabel('Variation of the diphoton mass ($m_{\gamma\gamma}$)')
plt.ylabel('Number of Events')
plt.legend()
plt.show()
```

# Visualization of flat distribution data.

```
Flat=pd.read_csv('flat_data.csv')

plt.figure(figsize = (12,5))
sns.distplot(Flat['flat'], color = 'skyblue', kde = False, label = 'Flat distribution')
plt.xlabel('Variation of the diphoton mass ($m_{\gamma\gamma}$)')
plt.ylabel('Number of Events')
plt.legend()
plt.show()
```
