## **We want to generate four random functions that are gaussian, exponential falling, uniform and poisson**.

3 small data sets with different seed will be generated and group into one data set for each type of file.

-  First, we want to find the seed of some numbers.

```
def digits(n):
  """to grab the digits of a number."""
  t = [int(i) for i in str(n)]
  return np.product(t)
```

```
# find the seed of numbers between 10000 and 25000.
def seeds(n):
  divisors = []
  factors = []

  for i in range(1,n):
    if n%i == 0:
      divisors.append(i)
  for i in divisors:
    if len(str(i))>1:
      factors.append(i)
  for i in factors:
    if i*digits(i)== n:
      return i
```

```
n=20216
print('seed of {} is: '.format(n), seeds(n))
```

```
n=20265
print('seed of {} is: '.format(n), seeds(n))
```

```
n=10071
print('seed of {} is: '.format(n), seeds(n))
```

- Random Gaussian distribution of Mean = 125 GeV, standard deviation = 3.5 GeV.

```
mu = 125
sigma =  9
```

```
n=20216
np.random.seed(722)
gauss1=np.random.normal(mu, sigma,n)
dict1={'gauss': gauss1}
df=pd.DataFrame(dict1)
df.to_csv(r'gauss1.csv', index = False) 
```

```
n = 20265
np.random.seed(1351)
gauss2=np.random.normal(mu, sigma,n)
dict2={'gauss': gauss2}
df = pd.DataFrame(dict2)
df.to_csv(r'gauss2.csv', index = False) 
```

```
n = 10071
np.random.seed(1119)
gauss3=np.random.normal(mu, sigma,n)
dict3 = {'gauss': gauss3}
df = pd.DataFrame(dict3)
df.to_csv(r'gauss3.csv', index = False) 
```

```
# Group the subdata into a data for the Gaussian and plot it.
data_1 = pd.read_csv('gauss1.csv')
data_2 = pd.read_csv('gauss2.csv')
data_3 = pd.read_csv('gauss3.csv')

gauss_data = pd.concat([data_1, data_2, data_3], axis = 0)
gauss_data.to_csv(r'gauss_data.csv', index = False) 

Gauss= pd.read_csv('gauss_data.csv')
Gauss.head()
```
